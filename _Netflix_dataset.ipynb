## Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head(2)
```

```python
import pandas as pd

# Assuming df is your DataFrame
df['date_added'] = df['date_added'].str.strip()  # Remove leading/trailing spaces
df['Date_N'] = pd.to_datetime(df['date_added'])

# Check for invalid dates
invalid_dates = df[df['Date_N'].isna()]
if not invalid_dates.empty:
    print("Invalid dates found:")
    print(invalid_dates)

df['Date_N'].fillna(pd.Timestamp('2027-01-01'))
```

```python
df['Date_N'].dt.year.value_counts()
```

```python
df['Date_N'].dt.year.value_counts().plot(kind='bar')
```
